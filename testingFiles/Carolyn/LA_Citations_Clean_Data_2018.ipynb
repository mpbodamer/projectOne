```python
# Import dependencies

import os
import csv
import pandas as pd

# Store filepath in a variable
    
file = "../Carolyn/Parking_Citations_in_FY_2018.csv"


# Read our Data file with the pandas library

citation_df = pd.read_csv(file, encoding="UTF-8", low_memory = False)    

# Preview Data Frame

citation_df.head()
```

```python
# Create new Data Frame with specific columns

citation_df_2 = citation_df[["Ticket number", "Issue Date", "Fine amount", "Latitude", "Longitude"]]

# Preview Data Frame

citation_df_2.head()
```

```python
# Drop any "NA" values

citation_df_2 = citation_df_2.dropna()

# Create a new column titled "Year" that takes only the year data from the "Issue Date" column

citation_df_2["Year"] = citation_df_2["Issue Date"].str[-4:]

# Filter the Data Frame to have only 2018 data

data_2018 = citation_df_2[citation_df_2["Year"] == "2018"]

# Preview Data Frame

data_2018.head()
```

```python
# Drop columns that we will not be using

final_df = data_2018.drop(columns={"Issue Date", "Year"})

# Preview Data Frame

final_df.head()
```

```python
# Save as CSV file

path = r"C:/Users/Carolyn/Desktop/projectOne/main/resources"

final_df.to_csv(os.path.join(path,r'clean_citation_data_2018.csv'))

# Get a final count of values

final_df.count()
```
